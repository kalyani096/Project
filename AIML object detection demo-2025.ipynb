```python
frozen_model = 'frozen_inference_graph.pb'
```

```python
model=cv2.dnn_DetectionModel(frozen_model,config_file)
```

```python
classLabels=[]

filename='labels.txt'

with open(filename,'rt') as fpt:
    classLabels= fpt.read().rstrip('\n').split('\n')
```

```python
print(classLabels)
```

```python
#thses are the diff categories of coco dataset
```

```python
print(len(classLabels))
```

```python
model.setInputSize(250,250)

model.setInputScale(1.0/127.5)

model.setInputMean((127.5,5,127,127.5))

model.setInputSwapRB(True)
```

```python
image=cv2.imread('son.jpg')

plt.imshow(image)
```

```python
ClassIndex,confidence,bbox=model.detect(image,confThreshold=0.5)
```

```python
print(ClassIndex)
```

```python
#label = classLabels[ClassInd - 1]  # Get the label string
#position = (boxes[0] + 10, boxes[1] + 40)
ClassIndex,confidence,bbox=model.detect(image,confThreshold=0.5)

font_scale =3

font=cv2.FONT_HERSHEY_PLAIN

for ClassInd, conf, boxes in zip(ClassIndex.flatten(), confidence.flatten(), bbox):
    label = classLabels[ClassInd - 1]
    cv2.rectangle(image,boxes,(250,0,0),2)
    cv2.putText(image, label, (boxes[0]+10, boxes[1]+40), font, font_scale, (0,255,0), 3)



```

```python
#Video
#cap = cv2.VideoCapture(1)
cap=cv2.VideoCapture('peels-george-morina-5330833.mp4')
if not cap.isOpened():
    cap = cv2.VideoCapture(0)
if not cap.isOpened():
    raise IOError('Cannot open the video')

font_scale = 3
font = cv2.FONT_HERSHEY_PLAIN

while True:
    ret, frame = cap.read()
    if not ret:
        break

    ClassIndex, confidence, bbox = model.detect(frame, confThreshold=0.55)
    print(ClassIndex)

    if len(ClassIndex) != 0:
        for ClassInd, conf, boxes in zip(ClassIndex.flatten(), confidence.flatten(), bbox):
            if ClassInd <= 80:
                label = classLabels[ClassInd - 1]  # Get label string
                position = (boxes[0] + 10, boxes[1] + 40)

                cv2.rectangle(frame, boxes, (250, 0, 0), 2)
                cv2.putText(frame, label, (boxes[0]+10, boxes[1]+40), font, font_scale, (0,255,0), 3)
                #cv2.putText(frame, label, position, font, font_scale, (0, 255, 0), 3)

    cv2.imshow('Obj detection', frame)

    if cv2.waitKey(2) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()

#fine Done
```
