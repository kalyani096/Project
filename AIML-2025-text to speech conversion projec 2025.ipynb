```python
from tkinter import *
from tkinter.ttk import Combobox

# Initialize window
root = Tk()
root.title('Simplilearn Text to Speech Converter')
root.geometry('1000x580+200+80')
root.configure(bg='#F7AC40')

# Header Frame
header_frame = Frame(root, bg='#14A7DD', width=1000, height=120)
header_frame.place(x=0, y=0)

# Logo
try:
    simplilearn_logo = PhotoImage(file='simpli.png')
    logo_label = Label(header_frame, image=simplilearn_logo, bg='#14A7DD')
    logo_label.image = simplilearn_logo  # Prevent garbage collection
    logo_label.place(x=20, y=10)
except:
    print("Logo image not found.")

# Title
Label(header_frame, text='Text to Speech Converter',
      font='TimesNewRoman 36 bold',
      bg='#14A7DD', fg='white').place(x=250, y=30)

# Text Input Box
text_box = Text(root, font='Calibri 23', bg='white',
                relief='solid', wrap=WORD, bd=0)
text_box.place(x=33, y=150, width=950, height=170)

# Voice Label & Selector
Label(root, text='Select Voice',
      font='TimesNewRoman 17 bold',
      bg='#F7AC40', fg='white').place(x=200, y=340)

gender_box = Combobox(root,
                      values=['Male', 'Female'],
                      state='readonly',
                      width=11,
                      font='Arial 12')
gender_box.place(x=350, y=340)
gender_box.set('Male')

# Speed Label & Selector
Label(root, text='Select Speed',
      font='TimesNewRoman 17 bold',
      bg='#F7AC40', fg='white').place(x=200, y=380)

speed_box = Combobox(root,
                     values=['Fast', 'Medium', 'Slow'],
                     state='readonly',
                     width=11,
                     font='Arial 12')
speed_box.place(x=350, y=380)
speed_box.set('Medium')

# Play Button
try:
    play_button = PhotoImage(file='playbutton.png')
    play_btn = Button(root, text='Play',
                      image=play_button,
                      compound=LEFT,
                      bg='white',
                      borderwidth='0.1c'
                      font='Arial 17 bold',
                      width=302)
    play_btn.image = play_button,
    play_btn.place(x=500, y=430)
except:
    print("Play button image not found.")

# Start GUI loop
root.mainloop()
```
