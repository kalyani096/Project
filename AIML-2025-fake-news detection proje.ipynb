```python
df_merge=pd.concat([df_fake,df_true],axis=0)
```

```python
df=df_merge.drop(['title','subject','date'],axis=1)
```

```python
df.isnull().sum()
```

```python
df=df.sample(frac=1)
```

```python
df.head()
```

```python
df.reset_index(inplace=True)
```

```python
df.drop(['index'],axis=1,inplace=True)
```

```python
def wordopt(text):
    text=text.lower()
    text=re.sub('\[.*?\]','',text)
    text=re.sub('\\W','',text)
    text=re.sub('http?://\S+|www\.\S+',' ' , text)
    text=re.sub('<.*?>+', ' ',text)
    text=re.sub('[%s]'  % re.escape(string.punctuation),' ',text)
    text=re.sub('\n',' ' , text)
    text=re.sub('\w*\d\w',' ' , text)

    return text
    
```

```python
df['text']=df['text'].apply(wordopt)
```

```python
x=df['text']

y=df['class']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=102)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer=TfidfVectorizer()

xv_train=vectorizer.fit_transform(x_train)

xv_test=vectorizer.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(xv_train,y_train)
```

```python
pred_lr=lr.predict(xv_test)
```

```python
lr.score(xv_test,y_test)
```

```python
print(classification_report(y_test,pred_lr))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree=DecisionTreeClassifier()

dtree.fit(xv_train,y_train)
```

```python
pred_dtree=dtree.predict(xv_test)
```

```python
dtree.score(xv_test,y_test)
```

```python
print(classification_report(y_test,pred_dtree))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
gbc=GradientBoostingClassifier()

gbc.fit(xv_train,y_train)
```

```python
pred_gbc=gbc.predict(xv_test)
```

```python
gbc.score(xv_test,y_test)
```

```python
print(classification_report(y_test,pred_gbc))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc=RandomForestClassifier()

rfc.fit(xv_train,y_train)
```

```python
pred_rfc=rfc.predict(xv_test)
```

```python
rfc.score(xv_test,y_test)
```

```python
print(classification_report(y_test,pred_rfc))
```

```python
def output_label(n):
    if n==1:
        return "Fake News"
    elif n==0:
        return "Not a Fake News"
        
```

```python
def manual_testing(news):
    testing_news={'text':[news]}
    new_df_test=pd.DataFrame(testing_news)
    new_df_test['text']=new_df_test['text'].apply(wordopt)
    new_x_test=new_df_test['text']
    new_xv_test=vectorizer.transform(new_x_test)
    pred_lr=lr.predict(new_xv_test)
    pred_rfc=rfc.predict(new_xv_test)
    pred_dtree=dtree.predict(new_xv_test)
    pred_gbc=gbc.predict(new_xv_test)

    return print("\n\n LR prediction : {}  \nDtree prediction : {} \n rfc prediction : {} \n gbc prediction :{}". format(output_label(pred_lr[0]),
                                                                                                                         output_label(pred_rfc[0]),
                                                                                                                         output_label(pred_gbc[0]),
                                                                                                                         output_label(pred_dtree[0])))
                
                 
    
```
