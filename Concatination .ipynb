```python
import pandas as pd
import numpy as np
```

```python
df1=pd.read_csv("north_america_2000_2010.csv")
df1
```

```python
df2=pd.read_csv("south_america_2000_2010.csv")
df2
```

```python
df1.head()
```

```python
df2.head()
```

```python
#data concatination col wise
df3=pd.concat([df1,df2],axis=1)
df3
```

```python
#concatination row wise
df3=pd.concat([df1,df2])
df3
```

```python
#data concatinated by index
df3=pd.concat([df1,df2],ignore_index=True)
df3
```

```python
#Identifying missing values in the dataset in the form of boolean format
df3.isnull()
```
