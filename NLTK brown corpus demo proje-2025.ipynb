```python
import nltk

# Step 1: Sentence Tokenization
sentences = nltk.sent_tokenize(data2)

# Step 2: Word Tokenization
tokenized_sentences = [nltk.word_tokenize(sentence) for sentence in sentences]

# Step 3: POS Tagging
tagged_sentences = [nltk.pos_tag(sentence) for sentence in tokenized_sentences]

# Step 4: Named Entity Chunking (binary=True groups all named entities under 'NE')
chunked_sentences = nltk.ne_chunk_sents(tagged_sentences, binary=True)

# Step 5: Recursive Entity Extraction
def extract_entity_names(tree):
    entity_names = []

    if hasattr(tree, 'label') and callable(tree.label):
        if tree.label() == 'NE':
            entity_names.append(' '.join([child[0] for child in tree]))
        else:
            for child in tree:
                entity_names.extend(extract_entity_names(child))
    return entity_names

# Step 6: Aggregate Entities
entity_names = []
for tree in chunked_sentences:
    entity_names.extend(extract_entity_names(tree))

# Step 7: Display Unique Entities
print(set(entity_names))
```
