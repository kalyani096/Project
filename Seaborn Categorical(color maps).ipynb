```python
#import libraries
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
#load dataset
df=sns.load_dataset('tips')
df.head()
```

```python
#HeatMap
fig=plt.figure(figsize=(5,4))
sns.heatmap(df.corr(),annot=True,fmt='0.2f',cmap='gist_rainbow')
```

```python
#Bar Plot
sns.barplot(x='sex',y='total_bill',data=df)
```

```python
#count plot
sns.countplot(x='day',hue='time',data=df)
```

```python
#box plot
sns.boxplot(x='total_bill',data=df)
```

```python
#box plot
sns.boxplot(x='sex',y='total_bill',palette='prism',data=df)
```

```python
#box plot
sns.boxplot(x='time',y='total_bill',hue='sex',palette='coolwarm',data=df)
plt.legend(loc=(0.4,0.8))
```

```python
#box plot
sns.boxplot(x='sex',y='total_bill',hue='day',palette='prism',data=df)
```

```python
#strip plot
sns.stripplot(x='day',y='total_bill',hue='sex',jitter=True,palette='gist_ncar',data=df)
```

```python
#swarm plot
sns.swarmplot(x='day',y='total_bill',hue='sex',palette='gist_ncar',split=True,data=df)
```

```python
#count plot
sns.set_style('whitegrid')
sns.countplot(x='sex',data=df)
```
