```python
from transformers import Wav2Vec2ForCTC,Wav2Vec2ForCTC,Wav2Vec2Tokenizer
```

```python
import numpy as np
```

```python
tokenizer=Wav2Vec2Tokenizer.from_pretrained("facebook/wav2vec2-base-960h")

model=Wav2Vec2ForCTC.from_pretrained("facebook/wav2vec2-base-960h")


                                        
```

```python
audio,sampling_rate=librosa.load("welcome.mp3",sr=16000)


```

```python
audio,sampling_rate
```

```python
import sys

sys.path.append("download")   
    #sys.path.append("/path/to/your/audio/directory")
import pygame
pygame.mixer.init()
sound = pygame.mixer.Sound("welcome.mp3")
sound.play()


```

```python
input_values=tokenizer(audio,return_tensors='pt').input_values

input_values
```

```python
logits=model(input_values).logits

logits
```

```python
predicted_ids=torch.argmax(logits,dim=-1)
```

```python
transciptions=tokenizer.decode(predicted_ids[0])
```
