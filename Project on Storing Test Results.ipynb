```python
#import pandas library
import pandas as pd
```

```python
#create a dataframe using as given in the project
data={"First_Name":["Jason","Molly","Tina","Jake","Army"],
     "Last_Name":["Miller","Jacobson",".","Milner","Cooze"],
     "Age":[42,52,36,24,73],
     "Pretest_Score":[4,24,31,".","."],
     "Posttest_Score":[24000,9400057,57,62,70]}
df=pd.DataFrame(data,columns=["First_Name","Last_Name","Age","Pretest_Score","Posttest_Score"])
df
```

```python
#export the dataframe as a file using to_csv method
df.to_csv("project.csv")
```
