```python
#import the libraries
import pandas as pd
pd.set_option('max_rows',200)
pd.set_option('max_columns',200)
import warnings
warnings.filterwarnings("ignore")
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Q1
#import the dataset as csv file
df=pd.read_csv("311_Service_Requests_from_2010_to_Present.csv")
df.head(10)
```

```python
#identify the top 10 rows
df.head(10)
```

```python
#to view the columns
df.columns
```

```python
#view the top 1 row of dataset
df.head(1)
```

```python
#identifying the missing values of dataset
df.isna().sum()
```

```python
#percentage of NAN values
(df.isna().sum()/len(df))*100
```

```python
#drop columns which are having NAN values high
df.drop(['School or Citywide Complaint' , 'Vehicle Type' , 'Taxi Company Borough' , 
         'Taxi Pick Up Location' , 'Garage Lot Name'] , axis = 1 , inplace = True)
```

```python
df.Agency.unique()
```

```python
#Q2
#CCreate a new column Request_closing_time by exploring the dataset
df["Created Date"]=pd.to_datetime(df["Created Date"])
df["Closed Date"]=pd.to_datetime(df["Closed Date"])
df["Request_Closing_Time"]=df["Closed Date"]-df["Created Date"]
df["Request_Closing_Time"]
```

```python
#converting time closed to minute scale
df["Request_Closing_Time_minutes"]=df["Request_Closing_Time"].dt.seconds/60
df['Request_Closing_Time_minutes']

```

```python
#Q3 - insight 1 for Q3
agen_compl=df.groupby('Agency')['Complaint Type'].value_counts().to_frame().rename(columns={'Complaint Type':'Count'}).sort_values('Count',ascending=False)
agen_compl.reset_index(inplace=True)



# Graph
fig=plt.figure(figsize=(10,8))
plt.grid()
plt.barh(agen_compl['Complaint Type'],agen_compl['Count'],  label='complaint count',
         color='#7FFF00',edgecolor='black', linewidth=1.2, alpha=0.8)
for i, v in enumerate(agen_compl['Count']):
    plt.text(v + 0.2, i + .1, str(v), color='blue', fontweight='bold')

plt.show()

#  The insight tells us the maximum complains are for Blocked Driveway and Minimum for Animal in Park
```

```python
#Q3
# insight 2 for Q3
desc_comp=df.groupby(['City','Descriptor'])['Complaint Type'].count().to_frame().sort_values('Complaint Type',ascending=False)
desc_comp.reset_index(inplace=True)
desc_comp2=desc_comp.iloc[:15]

fig=plt.figure(figsize=(8,8))
plt.pie( desc_comp2['Complaint Type'] ,# data
        labels = desc_comp2['Descriptor'] ,      
        autopct = '%0.0f%%' ,    # autopct enables you to display the percent value using Python string formatting. 
                                 #... 2f' , then for each pie wedge, the format string is '%. 2f' autopct shows percent weight of slice ,  we have to learn the string code
        shadow = True ,          # shadow displays the shadow of slices
        radius =  1,           # for radius of the circle
        textprops={'fontsize': 15,
                  'color':'black'} );  

plt.show()
```

```python
# insight 3 for Q3
city_descr=df.groupby(['City','Descriptor'])['Complaint Type'].value_counts().to_frame().rename(columns={'Complaint Type':'Count'}).sort_values('Count',ascending=False)

city_descr.reset_index(inplace=True)
city_descr
```

```python
#Q3
# insight 4 for Q3
city_complain=df.groupby('City')['Complaint Type'].count().to_frame().sort_values('Complaint Type', ascending=False).reset_index()

city_complain2=city_complain.iloc[:5]

fig=plt.figure(figsize=(8,8))
plt.pie( city_complain2['Complaint Type'] ,# data
        labels = city_complain2['City'] ,      
        autopct = '%0.0f%%' ,    # autopct enables you to display the percent value using Python string formatting. 
                                 #... 2f' , then for each pie wedge, the format string is '%. 2f' autopct shows percent weight of slice ,  we have to learn the string code
        shadow = True ,          # shadow displays the shadow of slices
        radius =  1,           # for radius of the circle
        textprops={'fontsize': 15,
                  'color':'black'} );  

plt.show()

# Pie chart showing top 5 cities with max. complaints
```
